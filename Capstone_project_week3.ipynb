```python
# importing library

import pandas as pd 

import requests 

from bs4 import BeautifulSoup  
```

Following codes are to Scrape the Wikipedia page, wrangle the data and save as data frame using beautifulsoup and Pandas

```python
# importing data from wiki

file_url="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
req = requests.get(file_url) 

# Using beautifulsoup

soup = BeautifulSoup(req.content,'lxml') 

# looking dor tables

table = soup.find_all('table')[0]  

df = pd.read_html(str(table)) 
```

```python
# converting to dataframes, printing the size 

nb=pd.DataFrame(df[0]) 


snb=nb.shape
col_nb=snb[1]
row_nb=snb[0]

print('Row: ',row_nb,)
print('Column: ',col_nb)
```
